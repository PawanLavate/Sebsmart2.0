```python
from flask import Flask, request, jsonify
import joblib
import os

app = Flask(__name__)

MODEL_PATH = "yield_estimator_model.pkl"
ENCODER_PATH = "yield_estimator_label_encoders.pkl"
SCALER_PATH = "yield_estimator_scaler.pkl"

# ---------- SAFE MODEL LOADING ----------
def safe_load(path, name):
    if not os.path.exists(path):
        print(f"[ERROR] {name} file not found at: {path}")
        return None
    
    try:
        return joblib.load(path)
    except Exception as e:
        print(f"[ERROR] Could not load {name}: {e}")
        return None

model = safe_load(MODEL_PATH, "Model")
label_encoders = safe_load(ENCODER_PATH, "Label Encoders")
scaler = safe_load(SCALER_PATH, "Scaler")

# If any file failed to load → stop server safely
if model is None or label_encoders is None or scaler is None:
    print("\n❌ Startup failed: Missing model or preprocessing files.\n")
    import sys
    sys.exit(1)


# ---------------- API ROUTES -----------------

@app.route("/", methods=["GET"])
def home():
    return jsonify({"message": "Apple Yield Prediction API Running"})


@app.route("/yield", methods=["POST"])
def predict():
    try:
        data = request.get_json()

        required_fields = [
            "District_name", "species", "growth_stage",
            "soil", "weather", "pesticides", "tree_count"
        ]

        missing = [f for f in required_fields if f not in data]
        if missing:
            return jsonify({"error": f"Missing fields: {missing}"}), 400

        # Encode categorical features
        for col in ["District_name", "species", "growth_stage", "soil", "weather", "pesticides"]:
            if data[col] not in label_encoders[col].classes_:
                return jsonify({"error": f"Invalid value '{data[col]}' for {col}"}), 400

            data[col] = label_encoders[col].transform([data[col]])[0]

        # Scale tree_count
        data["tree_count"] = scaler.transform([[data["tree_count"]]])[0][0]

        # Arrange features in model’s exact order
        features = [
            data["District_name"],
            data["species"],
            data["growth_stage"],
            data["soil"],
            data["weather"],
            data["pesticides"],
            data["tree_count"]
        ]

        prediction = model.predict([features])[0]

        return jsonify({
            "prediction": float(prediction),
            "status": "success"
        })

    except Exception as e:
        return jsonify({"error": str(e)}), 500


if __name__ == "__main__":
    # Prevent debugger reload crash
    app.run(host="0.0.0.0", port=5000, debug=False)
```
